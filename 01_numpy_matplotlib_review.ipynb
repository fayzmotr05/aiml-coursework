## Section 3: Dot Product (Core ML Operation)

**What this section is about:**  
The dot product implements the weighted sum used in perceptrons and neural networks.  
This is the most important NumPy operation for this course.

```python
x = np.array([2, 3])
w = np.array([0.5, -1.0])

z = np.dot(x, w)
print("Dot product:", z)
```

## Section 4: Step Activation Function

**What this section is about:**  
Activation functions turn numbers into decisions.  
Here we use a simple step function to simulate neuron firing.

```python
def step_function(z):
    return 1 if z >= 0 else 0

print(step_function(0.3))
print(step_function(-1.2))
```

## Section 5: Simple Perceptron Prediction

**What this section is about:**  
This section combines dot product, bias, and activation into a single prediction function.  
This is a complete single-neuron net.

image.png

```python
def predict(x, w, b):
    z = np.dot(x, w) + b
    return step_function(z)

x = np.array([2, 1])
w = np.array([1.0, -0.5])
b = -0.2

print("Prediction:", predict(x, w, b))
```

### ✏️ Exercise 5
Change the values of `w` and `b`.  
Observe how the prediction changes.

```python
w = np.array([3.0, -0.2])
b = -0.1

prediction = predict(x, w, b)
print("Prediction:", prediction)
```

## Section 6: Data Visualization with Matplotlib

**What this section is about:**  
Visualization helps us understand how data points are distributed.  
This section plots 2D data used for perceptron training.

```python
import matplotlib.pyplot as plt

X = np.array([[2,3], [1,1], [2,1], [3,2]])
y = np.array([1, 0, 0, 1])

plt.scatter(X[:,0], X[:,1], c=y)
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.title("Training Data")
plt.show()
```

## Section 7: Decision Boundary Preview

**What this section is about:**  
A perceptron learns a straight-line decision boundary.  
This section previews how weights and bias define that line.

```python
w = np.array([-1, -1])
b = 3.5

x_vals = np.linspace(0, 4, 100)
y_vals = -(w[0] * x_vals + b) / w[1]

plt.scatter(X[:,0], X[:,1], c=y)
plt.plot(x_vals, y_vals)
plt.title("Decision Boundary Preview")
plt.show()
```
