## 7. Manipulating Tensors (Basic Operations)

### Exercise 7.1: Perform basic operations
Create a tensor [10, 20, 30]. Add 5, subtract 10, multiply by 2, and divide by 10.

```python
tensor = torch.tensor([10, 20, 30])
# TODO: Add 5
added = tensor + 5
print(f"Added 5: {added}")

# TODO: Subtract 10
subtracted = tensor - 10
print(f"Subtracted 10: {subtracted}")

# TODO: Multiply by 2
multiplied = tensor * 2
print(f"Multiplied by 2: {multiplied}")

# TODO: Divide by 10
divided = tensor / 10
print(f"Divided by 10: {divided}")
```

### Exercise 7.2: Matrix multiplication
Create two tensors A (2x3) and B (3x2) with random values and perform matrix multiplication.

```python
# TODO: Create A and B
A = torch.rand(2, 3)
B = torch.rand(3, 2)

print(f"Tensor A:\n{A}")
print(f"Tensor B:\n{B}")

# TODO: Matrix mul using torch.mm
result_mm = torch.mm(A, B)
print(f"Result using torch.mm:\n{result_mm}")

# TODO: Matrix mul using @
result_at = A @ B
print(f"Result using @:\n{result_at}")
```

## 8. Tensor Aggregation

### Exercise 8.1: Find min, max, mean, sum
Create a tensor with values from 0 to 100 and find its min, max, mean, and sum.

```python
agg_tensor = torch.arange(0, 101)
# TODO: Min
min_val = agg_tensor.min()
print(f"Min: {min_val}")

# TODO: Max
max_val = agg_tensor.max()
print(f"Max: {max_val}")

# TODO: Mean (note: may need to cast to float)
mean_val = agg_tensor.type(torch.float32).mean()
print(f"Mean: {mean_val}")

# TODO: Sum
sum_val = agg_tensor.sum()
print(f"Sum: {sum_val}")
```

### Exercise 8.2: Argmin and argmax
Find the positions of the minimum and maximum values in the above tensor.

```python
# TODO: Argmin
argmin_val = agg_tensor.argmin()
print(f"Argmin: {argmin_val}")

# TODO: Argmax
argmax_val = agg_tensor.argmax()
print(f"Argmax: {argmax_val}")
```

## 9. Reshaping, Stacking, Squeezing, Unsqueezing

### Exercise 9.1: Reshape a tensor
Create a tensor of shape (10,) and reshape it to (2, 5).

```python
reshape_tensor = torch.arange(10)
print(f"Original tensor: {reshape_tensor}, Shape: {reshape_tensor.shape}")
# TODO: Reshape
reshaped = reshape_tensor.reshape(2, 5)
print(f"Reshaped tensor: {reshaped}, Shape: {reshaped.shape}")
```

### Exercise 9.2: Stack tensors
Create two tensors of shape (2, 2) and stack them vertically and horizontally.

```python
tensor1 = torch.rand(2, 2)
tensor2 = torch.rand(2, 2)
print(f"Tensor 1:\n{tensor1}")
print(f"Tensor 2:\n{tensor2}")

# TODO: Stack vertically (dim=0)
vertically_stacked = torch.vstack([tensor1, tensor2])
print(f"Vertically stacked:\n{vertically_stacked}")

# TODO: Stack horizontally (dim=1)
horizontally_stacked = torch.hstack([tensor1, tensor2])
print(f"Horizontally stacked:\n{horizontally_stacked}")
```

### Exercise 9.3: Squeeze and unsqueeze
Create a tensor of shape (1, 3, 1), squeeze it, then unsqueeze it back on dim=0.

```python
squeeze_tensor = torch.rand(1, 3, 1)
print(f"Original tensor: {squeeze_tensor}, Shape: {squeeze_tensor.shape}")

# TODO: Squeeze
squeezed = squeeze_tensor.squeeze()
print(f"Squeezed tensor: {squeezed}, Shape: {squeezed.shape}")

# TODO: Unsqueeze on dim=0
unsqueezed = squeezed.unsqueeze(dim=0)
print(f"Unsqueezed tensor: {unsqueezed}, Shape: {unsqueezed.shape}")
```

### Exercise 9.4: Permute
Create an image tensor (3, 100, 100) and permute it to (100, 100, 3).

```python
image_tensor = torch.rand(3, 100, 100)
print(f"Original image tensor shape: {image_tensor.shape}")
# TODO: Permute
permuted = image_tensor.permute(1, 2, 0)
print(f"Permuted image tensor shape: {permuted.shape}")
```

## 11. PyTorch and NumPy

### Exercise 11.1: NumPy to PyTorch
Create a NumPy array [1,2,3,4] and convert it to a PyTorch tensor.

### Exercise 13.2: Move tensor to GPU
Create a tensor and move it to the GPU if available. Then move it back to CPU and convert to NumPy.

```python
gpu_tensor = torch.tensor([10, 20, 30])
print(f"Original tensor: {gpu_tensor}, Device: {gpu_tensor.device}")

# TODO: Move to device
device = "cuda" if torch.cuda.is_available() else "cpu"
gpu_tensor = gpu_tensor.to(device)
print(f"Tensor on device: {gpu_tensor}, Device: {gpu_tensor.device}")

# TODO: Move back to CPU and to NumPy
cpu_numpy = gpu_tensor.cpu().numpy()
print(f"Tensor moved to CPU and converted to NumPy: {cpu_numpy}, Type: {type(cpu_numpy)}")
```
